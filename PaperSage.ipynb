```python
# Import necessary libraries
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.core.prompts import PromptTemplate
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.huggingface import HuggingFaceLLM
import torch
from huggingface_hub import login
```

```python
# Set up embedding model
print("Setting up embedding model...")
embed_model = HuggingFaceEmbedding(
    model_name="BAAI/bge-large-en-v1.5",
    trust_remote_code=True
)
```

```python
if use_hf_token:
  # First level of indentation (inside the if)
  hf_token = input("Enter your Hugging Face token: ")
  os.environ["HF_TOKEN"] = hf_token

  # Login to Hugging Face with the token
  login(token=hf_token)

  # Use a small model that can run on Colab
  model_name = "google/gemma-2b"
  try:
    # Second level of indentation (inside the try)
    llm = HuggingFaceLLM(
        # Third level of indentation (for parameters)
        model_name=model_name,
        tokenizer_name=model_name,
        context_window=2048,
        max_new_tokens=256,
        generate_kwargs={"temperature": 0.1, "do_sample": False},
        device_map="auto"
    )
  except Exception as e:
    # Second level of indentation (inside the except)
    print(f"Error loading Gemma model: {e}")
    print("Falling back to TinyLlama model...")
    model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
    llm = HuggingFaceLLM(
        # Third level of indentation (for parameters)
        model_name=model_name,
        tokenizer_name=model_name,
        context_window=2048,
        max_new_tokens=256,
        generate_kwargs={"temperature": 0.1, "do_sample": False},
        device_map="auto"
    )
else:
  # First level of indentation (inside the else)
  print("Using a publicly available model...")
  model_name = "distilbert-base-uncased"  # Use a more accessible model
  llm = HuggingFaceLLM(
      # Second level of indentation (for parameters)
      model_name=model_name,
      tokenizer_name=model_name,
      context_window=2048,
      max_new_tokens=256,
      generate_kwargs={"temperature": 0.1, "do_sample": False},
      device_map="auto"
  )
```

```python
# Set global settings
Settings.embed_model = embed_model
Settings.llm = llm
```

```python
# Verify test-dir exists or upload PDF
pdf_dir = None
if os.path.exists('test-dir'):
  pdf_dir = 'test-dir'
  print(f"Using test-dir found at {os.path.abspath(pdf_dir)}")
  print(f"Files in test-dir: {os.listdir(pdf_dir)}")
else:
  # Prompt user to upload PDFs if no directory is found
  print("test-dir not found. Please upload your PDF files...")
  uploaded = files.upload()
  pdf_dir = tempfile.mkdtemp()
  for filename, content in uploaded.items():
    if filename.endswith('.pdf'):
      with open(os.path.join(pdf_dir, filename), 'wb') as f:
        f.write(content)

# Load documents
print(f"Loading documents from {pdf_dir}...")
loader = SimpleDirectoryReader(
    input_dir=pdf_dir,
    required_exts=[".pdf"],
    recursive=True
)
docs = loader.load_data()
print(f"Loaded {len(docs)} document(s)")
```

```python
# Create vector index
print("Creating vector index...")
index = VectorStoreIndex.from_documents(docs)
```

```python
# Define custom prompt template
qa_prompt_tmpl_str = (
    "Context information is below.\n"
    "---------------------\n"
    "{context_str}\n"
    "---------------------\n"
    "Given the context information above I want you to think step by step to answer the query in a crisp manner, incase case you don't know the answer say 'I don't know!'.\n"
    "Query: {query_str}\n"
    "Answer: "
)

qa_prompt_tmpl = PromptTemplate(qa_prompt_tmpl_str)

# Create query engine
print("Setting up query engine...")
query_engine = index.as_query_engine(streaming=True, similarity_top_k=4)
query_engine.update_prompts({"response_synthesizer:text_qa_template": qa_prompt_tmpl})
```

```python

# Test query
print("\nTest Query: What is PaperSage?")
response = query_engine.query("What is PaperSage?")
print(f"Response: {response}")
```

```python
import os
import pickle
from pyngrok import ngrok

# 1) Create the Streamlit app file
with open('app_hf.py', 'w') as f:
    f.write('''
import streamlit as st
import os
import pickle
import torch

# Streamlit page config
st.set_page_config(
    page_title="PaperSage: Chat with your Documents",
    layout="centered"
)

# Safe CUDA memory config
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

query_engine_path = 'query_engine.pkl'
query_engine = None

if os.path.exists(query_engine_path):
    try:
        with open(query_engine_path, 'rb') as f:
            query_engine = pickle.load(f, map_location='cpu')  # ✅ Force CPU loading
        st.success("Successfully loaded the query engine (CPU only)!")
    except Exception as e:
        st.error(f"Error loading query engine: {e}")
        st.warning("The query engine file exists but couldn't be loaded properly.")
else:
    st.error("Query engine not found!")
    st.warning("Please run the index & query engine creation cells, then re-run.")
    st.stop()
# Chat memory
if "messages" not in st.session_state:
    st.session_state.messages = []

# Replay old messages
for msg in st.session_state.messages:
    with st.chat_message(msg["role"]):
        st.markdown(msg["content"])

# User input handling
if query_engine and (prompt := st.chat_input("Ask something about your documents...")):
    st.session_state.messages.append({"role": "user", "content": prompt})
    with st.chat_message("user"):
        st.markdown(prompt)

    with st.chat_message("assistant"):
        msg_placeholder = st.empty()
        result = ""
        try:
            response = query_engine.query(prompt)
            result = str(response)
        except Exception as e:
            result = f"Error: {str(e)}"
        msg_placeholder.markdown(result)

    st.session_state.messages.append({"role": "assistant", "content": result})
''')

print("Streamlit app code saved to app_hf.py!")
```

```python
# 2) Define a helper to remove unpickleable attributes
def remove_unpickleable_attributes(obj):
    if not hasattr(obj, "__dict__"):
        return
    keys = list(obj.__dict__.keys())
    for attr in keys:
        val = getattr(obj, attr, None)
        try:
            pickle.dumps(val)
        except:
            print(f"Removing unpickleable attr: {attr}")
            delattr(obj, attr)

# 3) Save the query engine if it exists
if 'query_engine' in globals() or 'query_engine' in locals():
    print("Saving query_engine to 'query_engine.pkl'...")
    remove_unpickleable_attributes(query_engine)
    with open('query_engine.pkl', 'wb') as f:
        pickle.dump(query_engine, f)
    print("Query engine saved!")
elif 'index' in globals() or 'index' in locals():
    print("query_engine not found, but index is found. Creating query engine...")
    query_engine = index.as_query_engine(similarity_top_k=4)
    remove_unpickleable_attributes(query_engine)
    with open('query_engine.pkl', 'wb') as f:
        pickle.dump(query_engine, f)
    print("Index -> query_engine created and saved!")
else:
    print("ERROR: No query_engine or index found. Please build them first!")
```
